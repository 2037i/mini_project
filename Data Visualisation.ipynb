## Import TV Only Dataset

```python
animelist = pd.read_csv('DataSets/Cleaned data/tv_only_success_metric.csv')
pd.set_option('display.max_columns', None)
animelist.head()
```

```python
animelist.describe()
```

```python
numeric_data = pd.DataFrame(animelist[['episodes', 'year','days_aired','duration_minutes','Average Rank of Related','success']])
numeric_data.describe().round(2)
```

## Distributions of all numeric variables
---

```python
# Draw the distributions of all variables
f, axes = plt.subplots(numeric_data.shape[1], 3, figsize=(18, 25))

count = 0
for var in numeric_data:
    sb.boxplot(data = numeric_data[var], orient = "h", ax = axes[count,0])
    sb.histplot(data = numeric_data[var], ax = axes[count,1])
    sb.violinplot(data = numeric_data[var], orient = "h", ax = axes[count,2])
    count += 1
```

## Correlation of numeric variables
---

```python
# Correlation Matrix
numeric_corr = numeric_data.corr()
```

```python
# Using Seaborn version 13.0 solves issue of correlation value not showing
plt.figure(figsize=(16, 6))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(numeric_corr, dtype=np.bool_))
heatmap = sb.heatmap(numeric_corr, mask=mask, vmin=-1, vmax=1, annot=True, cmap='RdBu')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

### Pair Plots 
---

```python
# Draw pairs of variables against one another
g = sb.pairplot(data = numeric_data, corner=True)
for ax in g.diag_axes: ax.set_visible(False)
```

# Overview of Genre & Related Sections

## For all anime

### Genre

```python
genre_list = [col for col in animelist.columns if 'Genre_' in col]
related_list = [col for col in animelist.columns if 'Related_' in col]
```

```python
def get_count(list, df):   
    list_count=df[list].apply(pd.Series.value_counts).transpose()
    list_count['percentage'] = (list_count[1]/(list_count[0]+list_count[1])*100).round(2)
    list_count= list_count.sort_values(by='percentage', ascending=False)
    list_count=list_count.reset_index()
    return list_count
    
genre_count = get_count(genre_list, animelist)
genre_count
```

```python
# Create the bar plot
sb.catplot(x='percentage', y='index', kind='bar', data=genre_count, height=15, aspect=1.2)
plt.xlabel('Percentage')
plt.ylabel('Genre')
plt.title('Percentage of Each Genre')
plt.show()
```

### Related

```python
def get_related_count(related_list, list):
    related_count = get_count(related_list, list)
    related_count_extra = related_count.drop(columns=['index',0,1,'percentage'])
    related_count['>1'] = related_count[related_count_extra.columns].sum(axis=1)
    related_count = related_count.drop(columns=related_count_extra.columns)
    
    related_count['percentage']=((related_count[1]+related_count['>1'])/(related_count['>1']+related_count[0]+related_count[1])*100).round(2)
    related_count= related_count.sort_values(by='percentage', ascending=False)
    
    related_count['percentage >1']=(related_count['>1']/(related_count['>1']+related_count[0]+related_count[1])*100).round(2)
    return related_count

related_count = get_related_count(related_list, animelist)
related_count
```

```python
# Create the bar plot
sb.catplot(x='percentage', y='index', kind='bar', data=related_count, height=10, aspect=1.2)
plt.xlabel('Percentage')
plt.ylabel('Related')
plt.title('Percentage of Each Related Type')
plt.show()
```

## Looking at members VS popularity

```python
df = pd.read_csv('DataSets/Cleaned data/data_cleaned.csv')
```

```python
sorted_df_pop = df.sort_values(by='popularity', ascending=True)

top_30_pop = sorted_df_pop.head(30)[['title', 'popularity']]
top_30_pop
```

```python
# Sort DataFrame by members
sorted_df_members = df.sort_values(by='members', ascending=False)

top_30_members = sorted_df_members.head(30)[['title', 'members']]
top_30_members
```

### Conclusion about popularity VS members

From the 2 cells, we can conclude that 'popularity' is actually a ranking of the 'members' column such that the show with most 'members' has 'poplarity'=1, the show with 2nd most 'members' has 'popularity'=2, etc.

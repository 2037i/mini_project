```python
numeric_data = pd.DataFrame(animelist[['episodes', 'year','days_aired','duration_minutes','Average Rank of Related','success']])
numeric_data.describe().round(2)
```

# Distributions of all numeric variables
---

```python
# Draw the distributions of all variables
f, axes = plt.subplots(numeric_data.shape[1], 3, figsize=(18, 25))

def remove_outliers_zscore(data, threshold=3):
    # Remove outliers with zscore method
    z_scores = np.abs((data - np.mean(data)) / np.std(data))
    return data[z_scores < threshold]
    
count = 0
for var in numeric_data:
    numeric_data[var]=remove_outliers_zscore(numeric_data[var])
    sb.boxplot(data = numeric_data[var], orient = "h", ax = axes[count,0])
    sb.histplot(data = numeric_data[var], ax = axes[count,1])
    sb.violinplot(data = numeric_data[var], orient = "h", ax = axes[count,2])
    count += 1
```

We have removed outliers. It is evident that most anime follow a fixed format with about ~25min, airing for 100-200 days (between3-6 months) and 12-24 episodes.

# Correlation of numeric variables
---

```python
# Correlation Matrix
numeric_corr = numeric_data.corr()
```

```python
# Using Seaborn version 13.0 solves issue of correlation value not showing
plt.figure(figsize=(16, 6))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(numeric_corr, dtype=np.bool_))
heatmap = sb.heatmap(numeric_corr, mask=mask, vmin=-1, vmax=1, annot=True, cmap='RdBu')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

Correlation seems pretty weak when comparing individually. We mainly care about how different variables correlate to success.   
Days aired and episodes are strongly correlated by definition (more episodes -> more days to air)

# Overview of Genre & Related Sections

```python
# Assuming animelist is your DataFrame containing the necessary columns including 'popularity' and 'title' and genre columns
df = pd.DataFrame(animelist)

sorted_df = df.sort_values(by='success', ascending=True)
```

```python
# Iterate over rows using apply()
print("A random selection of 20 successful animes")
success1 = sorted_df.loc[sorted_df['success'] == 1]
success1.head(20).apply(lambda row: print(f"{row['title']} - Genre: {', '.join([col for col in sorted_df.columns if row[col] == 1 and 'Genre_' in col])}"), axis = 1)
```

## Genre

```python
genre_list = [col for col in animelist.columns if 'Genre_' in col]
related_list = [col for col in animelist.columns if 'Related_' in col]
```

### In All Anime

```python
def get_count(list, df):   
    list_count=df[list].apply(pd.Series.value_counts).transpose()
    list_count['percentage'] = (list_count[1]/(list_count[0]+list_count[1])*100).round(2)
    list_count= list_count.sort_values(by='percentage', ascending=False)
    list_count=list_count.reset_index()
    return list_count
    
genre_count = get_count(genre_list, animelist)
genre_count
```

```python
# Create the bar plot
sb.catplot(x='percentage', y='index', kind='bar', data=genre_count, height=15, aspect=1.2)
plt.xlabel('Percentage')
plt.ylabel('Genre')
plt.title('Percentage of Each Genre')
plt.show()
```

### Looking at Successful Anime

```python
success1_count = get_count(genre_list, success1)
success1_count
```

```python
# Create the bar plot
sb.catplot(x='percentage', y='index', kind='bar', data=success1_count, height=15, aspect=1.2)
plt.xlabel('Percentage')
plt.ylabel('Genre')
plt.title('Percentage of Each Genre')
plt.show()
```

### Findings
Looking at the genre distributions, we can see that there are a few genres that dominate the chart, such as comedy appearing in more than 50% of all shows, followed by action and adventure. We know that these are typical popular genres.

However, looking only at the successful anime. Comedy is still the leading genre, but Drama, Romance, Supernatural, School sees huge advances. This can possibly be an indicator that they are less common, yet more likely to be successful.

## Related

### In All Anime

```python
def get_related_count(related_list, list):
    related_count = get_count(related_list, list)
    related_count_extra = related_count.drop(columns=['index',0,1,'percentage'])
    related_count['>1'] = related_count[related_count_extra.columns].sum(axis=1)
    related_count = related_count.drop(columns=related_count_extra.columns)
    
    related_count['percentage']=((related_count[1]+related_count['>1'])/(related_count['>1']+related_count[0]+related_count[1])*100).round(2)
    related_count= related_count.sort_values(by='percentage', ascending=False)
    
    related_count['percentage >1']=(related_count['>1']/(related_count['>1']+related_count[0]+related_count[1])*100).round(2)
    return related_count

related_count = get_related_count(related_list, animelist)
related_count
```

```python
# Create the bar plot
sb.catplot(x='percentage', y='index', kind='bar', data=related_count, height=10, aspect=1.2)
plt.xlabel('Percentage')
plt.ylabel('Related')
plt.title('Percentage of Each Related Type')
plt.show()
```

### Looking At Successful Anime

```python
success1_related_count = get_related_count(related_list, success1)
success1_related_count
```

```python
# Create the bar plot
sb.catplot(x='percentage', y='index', kind='bar', data=success1_related_count, height=10, aspect=1.2)
plt.xlabel('Percentage')
plt.ylabel('Related')
plt.title('Percentage of Each Related Type')
plt.show()
```

### Findings & Comparisons
Similarly, here are the related media. A related media for an anime can be a manga or movie adaptation, side story, or sequel/prequel. Comparing between the general dataset and only the successful shows, we can see that adaptations and sequels seem to be very important as its proportion jumped by alot when only looking at the set of successful shows.

# Looking at Format

## Season

```python
season_df = animelist[['start_season', 'success']]
season_df = season_df.loc[season_df['start_season']!='0']
result = pd.DataFrame(season_df.groupby('start_season')['success'].value_counts(normalize=True).mul(100).round(2))
result = result.sort_values(by=['start_season', 'success'], ascending=True)

sorted_result = result.sort_index(level='success', ascending=True)
sorted_result
```

```python
# Plotting
plt.figure(figsize=(10, 10)) 
sorted_result.plot(kind='bar', stacked=True)
plt.xlabel('Season')
plt.ylabel('Percentage of Success')
plt.title('Percentage of Success in Each Season')
plt.legend(title='Success Level')
plt.xticks(rotation=45)
plt.show()
```

Fall and Summer have a slight edge over Spring & Winter. Note that the distribution is skewed and its only worth comparing between seasons in each success bracket

## Studio

```python
studio_df = animelist[['studio', 'success']]
studio_df['studio'].value_counts()
```

```python
result = pd.DataFrame(studio_df.groupby('studio')['success'].value_counts(normalize=True).mul(100).round(2))
result
```

There are too many studios (unlike seasons), so we will just take top 20 studios with most success <2 to look at generally successful studios

```python


# Filter the DataFrame to include only success levels less than 2
df_success_less_than_2 = studio_df[studio_df['success'] < 2]

# Calculate the percentage of success levels less than 2 for each season
success_less_than_2_percentage = df_success_less_than_2.groupby('studio').size() / studio_df.groupby('studio').size() * 100

# Sort the seasons based on the percentage of success levels less than 2
sorted_seasons = pd.DataFrame(success_less_than_2_percentage.sort_values(ascending=False)).head(20).reset_index()
sorted_seasons = sorted_seasons.rename(columns={0: "percent"})

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(sorted_seasons['studio'], sorted_seasons['percent'], color='skyblue')
plt.xlabel('Percentage of Successful Outcomes')
plt.ylabel('Studios')
plt.title('Percentage of Generally Succesful (Success <2) Outcomes by Studio')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest percentage at the top
plt.show()
```

As expected, well known studios lik Kyoto Animation, Bones, UFOTABLE have good success rates. Though this may not be representative for small studios due to the large number of shows under it.

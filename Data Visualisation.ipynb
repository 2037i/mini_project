### Essential Libraries

Let us begin by importing the essential Python Libraries.

> NumPy : Library for Numeric Computations in Python  
> Pandas : Library for Data Acquisition and Preparation  
> Matplotlib : Low-level library for Data Visualization  
> Seaborn : Higher-level library for Data Visualization  

```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics

```

## Import cleaned Data
---

```python
animelist = pd.read_csv('DataSets/Cleaned data/tv_only_success_metric.csv')
animelist.head()
```

```python
animelist.describe()
```

```python
numeric_data = pd.DataFrame(animelist[['episodes', 'score', 'scored_by', 'rank', 'popularity', 'members', 'favorites', 'duration_minutes']])
numeric_data.describe().round(2)
```

## Distributions of all numeric variables
---

```python
# Draw the distributions of all variables
f, axes = plt.subplots(3, 3, figsize=(18, 25))

count = 0
for var in numeric_data:
    sb.boxplot(data = numeric_data[var], orient = "h", ax = axes[count,0])
    sb.histplot(data = numeric_data[var], ax = axes[count,1])
    sb.violinplot(data = numeric_data[var], orient = "h", ax = axes[count,2])
    count += 1
```

## Correlation of all variables
---

```python
# Correlation Matrix
print(numeric_data.corr())

# Heatmap of the Correlation Matrix
f = plt.figure(figsize=(10, 10))
sb.heatmap(numeric_data.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 18}, cmap = "RdBu")
```

### Pair Plots 
---

```python
# Draw pairs of variables against one another
sb.pairplot(data = numeric_data)
```

```python
## score vs number of scores
f = plt.figure(figsize=(16, 8))
plt.scatter(numeric_data['score'], numeric_data['members'])

# Set labels and title
plt.xlabel('Overall Score')
plt.ylabel('Number of Scores')
plt.title('Anime Score vs. Number of Scores')
plt.show()
```

## Top 50 animes list (popularity)
---
    
    Base on popularity and compare whether member.

```python
# Assuming animelist is your DataFrame containing the necessary columns including 'popularity' and 'title' and genre columns
df = pd.DataFrame(animelist)
df = df[df['popularity'] != 0]

# Sort DataFrame by popularity
sorted_df = df.sort_values(by='popularity', ascending=True)

# Get the top 10 titles by popularity
top_50 = sorted_df.head(50)

print("Top 50 titles by popularity:")
print(top_50[['title', 'popularity']])
```

```python
# Iterate over rows using apply()
print("Top 50 titles by popularity with genre:")
print()
top_50.apply(lambda row: print(f"{row['popularity']}: {row['title']} - Genre: {', '.join([col for col in sorted_df.columns if row[col] == 1 and 'Genre_' in col])}"), axis = 1)
```

### Comparing to the top 10 anime what are the genre is popular
---

```python
top_50_genre_counts = top_50[['Action', 'Adventure', 'Cars', 'Comedy', 'Dementia', 'Demons', 'Drama', 'Ecchi', 'Fantasy', 'Game', 'Harem', 'Hentai', 'Historical', 'Horror', 'Josei', 'Kids', 'Magic', 'Martial Arts', 'Mecha', 'Military', 'Music', 'Mystery', 'NA', 'Parody', 'Police', 'Psychological', 'Romance', 'Samurai', 'School', 'Sci-Fi', 'Seinen', 'Shoujo', 'Shoujo Ai', 'Shounen', 'Shounen Ai', 'Slice of Life', 'Space', 'Sports', 'Super Power', 'Supernatural', 'Thriller', 'Vampire', 'Yaoi', 'Yuri']].sum()

top_50_genre_counts = top_50_genre_counts.reset_index()
top_50_genre_counts.columns = ['Genre', 'Count']

# Sort the DataFrame by count
top_50_genre_counts = top_50_genre_counts.sort_values(by='Count', ascending=False)
top_50_genre_counts = top_50_genre_counts[top_50_genre_counts['Count'] > 0]

print("Genres in the top 10 most popular anime:")
print(top_50_genre_counts)
print()

# Create the bar plot
sb.catplot(x='Count', y='Genre', kind='bar', data=top_50_genre_counts, height=10, aspect=2)
plt.xlabel('Count')
plt.ylabel('Genre')
plt.title('Count of Each Genre')
plt.show()
```
